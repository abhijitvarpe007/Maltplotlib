### Ploting Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes=['Python','R','AI','ML','DS']
Class1_Students=[10,10,20,25,10]  #out of 100 student in each class
class2_Students=[40,5,20,20,10]
class3_Students=[35,5,30,15,15]
```

```python
plt.bar(classes,Class1_Students)
plt.xlabel("Class")
plt.ylabel("No Of Students")
plt.show()
```

```python
plt.barh(classes,Class1_Students)
plt.xlabel("No Of Students")
plt.ylabel("Class")
plt.show()
```

```python
plt.bar(classes,Class1_Students,width=0.2)
plt.xlabel("Class")
plt.ylabel("No Of Students")
plt.show()
```

```python
plt.bar(classes,Class1_Students,width=0.5,align='edge')
plt.xlabel("Class")
plt.ylabel("No Of Students")
plt.show()
```

```python
plt.bar(classes,Class1_Students,width=0.5,align='edge',color='m',edgecolor='r',linewidth=3,alpha=0.3)
plt.xlabel("Class")
plt.ylabel("No Of Students")
plt.show()
```

```python
plt.bar(classes,Class1_Students,width=0.5,align='edge',color='m',edgecolor='r',linewidth=2,
        linestyle='-.',label='Class 1 Student')
plt.xlabel("Class")
plt.ylabel("No Of Students")
plt.legend()
plt.show()
```

```python
plt.bar(classes,Class1_Students,width=0.5,align='edge',color='m',edgecolor='r',linewidth=2,
        linestyle='-.',label='Class 1 Student',visible=False)
plt.xlabel("Class")
plt.ylabel("No Of Students")
plt.legend()
plt.show()
```

```python
style.use('ggplot')
plt.figure(figsize=(12,7))
plt.bar(classes,Class1_Students,width=0.5,align='edge',color='m',edgecolor='r',linewidth=2,
        linestyle='-.',label='Class 1 Student')
plt.xlabel("Class")
plt.ylabel("No Of Students")
plt.legend()
plt.show()
```

```python
plt.barh(classes,Class1_Students,align='edge',color='m',edgecolor='r',linewidth=2,
        linestyle='-.',label='Class 1 Student')
plt.title('Bar Chart of Class')
plt.xlabel("Class")
plt.ylabel("No Of Students")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,7))
plt.bar(classes,Class1_Students,width=0.2,label='Class 1 Student')
plt.bar(classes,class2_Students,width=0.2,label='Class 2 Student')

plt.title('Bar Chart of Class',fontsize=15)
plt.xlabel("Classes",fontsize=15)
plt.ylabel("No Of Students",fontsize=15)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,7))
classes_index=np.arange(len(classes))
width=0.2
plt.bar(classes_index,Class1_Students,width,label='Class 1 Student')
plt.bar(classes_index+width,class2_Students,width,label='Class 2 Student')
plt.bar(classes_index+width+width,class3_Students,width,label='Class 3 Student')

plt.xticks(classes_index+width,classes,rotation=20)       # for class name display
 
plt.title('Bar Chart of Class',fontsize=15)
plt.xlabel("Classes",fontsize=15)
plt.ylabel("No Of Students",fontsize=15)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,7))
classes_index=np.arange(len(classes))
width=0.2
plt.barh(classes_index,Class1_Students,width,label='Class 1 Student')
plt.barh(classes_index+width,class2_Students,width,label='Class 2 Student')
plt.barh(classes_index+width+width,class3_Students,width,label='Class 3 Student')

plt.yticks(classes_index+width,classes,rotation=20)       # for class name display
 
plt.title('Bar Chart of Class',fontsize=15)
plt.ylabel("Classes",fontsize=15)
plt.xlabel("No Of Students",fontsize=15)
plt.legend()
plt.show()
```
