### Pie Chart

```python
import matplotlib.pyplot as plt
```

```python
classes=['Python','R','AI','ML','DS']
class1_Students=[45,15,35,25,30]
explode=[0.03,0,0.2,0,0]
colors=['c','b','r','y','m']
```

```python
plt.pie([1])
plt.show()
```

```python
plt.pie(class1_Students,labels=classes)
plt.show()
```

```python
plt.pie(class1_Students,labels=classes,explode=explode,colors=colors)
plt.show()
```

```python
plt.pie(class1_Students,labels=classes,explode=explode,colors=colors,autopct='%0.2f%%')
plt.show()
```

```python
plt.pie(class1_Students,labels=classes,explode=explode,colors=colors,autopct='%0.2f%%',
       shadow=True,radius=1.4)
plt.show()
```

```python
plt.pie(class1_Students,labels=classes,explode=explode,colors=colors,autopct='%0.2f%%',
       shadow=True,radius=1.4,startangle=90)
plt.show()
```

```python
plt.pie(class1_Students,labels=classes,explode=explode,colors=colors,autopct='%0.2f%%',
       shadow=True,radius=1.4,startangle=270)
plt.show()
```

```python
textprops={'fontsize':15}
plt.pie(class1_Students,labels=classes,explode=explode,colors=colors,autopct='%0.2f%%',
       shadow=True,radius=1.4,startangle=270,textprops=textprops)
plt.show()
```

```python
wedgeprops={'linewidth':4,"width":1,'edgecolor':'k'}
plt.pie( class1_Students,labels=classes,explode=explode,colors=colors,autopct="0.2f%%",pctdistance=0.6,
    shadow=True,labeldistance=1.4,startangle=270,radius=1.1,counterclock=True,wedgeprops=wedgeprops,
       textprops=textprops,center=(2,3),frame=True,rotatelabels=True)
plt.show()
```

```python
wedgeprops={'linewidth':4,"width":2,'edgecolor':'k'}
plt.pie( class1_Students,labels=classes,explode=explode,colors=colors,autopct="0.2f%%",pctdistance=0.6,
    shadow=True,labeldistance=1.4,startangle=270,radius=1.1,counterclock=True,wedgeprops=wedgeprops,
       textprops=textprops,center=(2,3),frame=True,rotatelabels=True)
plt.show()
```

```python
plt.figure(figsize=(8,4))
wedgeprops={'linewidth':4,"width":2,'edgecolor':'k'}
plt.pie( class1_Students,labels=classes,explode=explode,colors=colors,autopct="0.2f%%",pctdistance=0.6,
    shadow=True,labeldistance=1.4,startangle=270,radius=1.1,counterclock=True,wedgeprops=wedgeprops,
       textprops=textprops,center=(2,3),frame=True,rotatelabels=True)
plt.show()
```

```python
textprops={'fontsize':15}
plt.pie(class1_Students,labels=classes,explode=explode,colors=colors,autopct='%0.2f%%',
       shadow=True,radius=1.4,startangle=270,textprops=textprops)
plt.legend()
plt.show()
```
