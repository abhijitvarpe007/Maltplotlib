### Scatter Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import style
```

```python
df_google=pd.read_csv("googleplaystore.csv",nrows=1000)
df_google.shape
```

```python
df_google.head()
```

```python
df_google.dtypes
```

```python
x=df_google['Rating']
y=df_google['Reviews']
```

```python
plt.scatter(x,y)
plt.title("Google Play Store Apps")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.figure(figsize=(12,7))
plt.scatter(x,y,c='m',marker='*',s=100)
plt.title("Google Play Store Apps")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.figure(figsize=(12,7))
plt.scatter(x,y,c='m',marker='*',s=100,alpha=0.5)
plt.title("Google Play Store Apps")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.figure(figsize=(12,7))
plt.scatter(x,y,c='m',marker='*',s=100,alpha=0.5,linewidth=4,edgecolors='g')
plt.title("Google Play Store Apps")
plt.xlabel("Rating
plt.ylabel("Reviews")
plt.show()
```

```python
plt.figure(figsize=(12,7))
plt.scatter(x,y,c='m',marker='*')
plt.scatter(x,df_google['Installs'],c='r')
plt.title("Google Play Store Apps")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```
